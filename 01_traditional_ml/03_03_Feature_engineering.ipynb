# Feature Engineering - 특성공학

## 1.  Feature Engeering 이란 
### 1) 좋은 Feature 란 ?
- `Target`과의 높은 관련성 
- `prediction` 시점에 알 수 있음 
- `numberic` 
- 충분한 `example` 수 
- 전문가의 도메인 지식 활용 가능 

### 2) 머신러닝을 위한 Feature Engineering
- 결측치 (Missing Value) 처리
- 데이터 형식화 (Data Formatting)
- 편향 (skewness) 완화
- 데이터 정규화 (Normalization)
- Binning(구간화)
- 범주형 데이터 수치화 (Categorical Data Encoding)
  - numeric 로 변경 
  - ordinal category(순서가 있는 카테고리)
  - nominal category (순서/크기가 없는 feature)
